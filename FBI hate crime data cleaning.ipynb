## TIME SERIES DATA : hate crime, year, month

### 1. All race

```python
#make data that only has state, data_year, month left
all_race = race_hc_s[['DATA_YEAR', 'INCIDENT_MONTH']].copy()
all_race.index = range(len(all_race))
all_race
```

```python
#for each year, aggregate by month
#first, create 1991 data as base + test run for loop for month
yearL=pd.unique(all_race["DATA_YEAR"])
i=0
all_race_ts = pd.DataFrame(columns=['year','month','hate_crime']) #create data frame for each year with 12 months
l=all_race.loc[all_race['DATA_YEAR']==yearL[i], "INCIDENT_MONTH"]
#month=1
for month in range(1,13):
    all_race_ts.loc[month-1] = [yearL[i], month, sum(l==month)]
all_race_ts
```

```python
yearL=yearL[1:30] #list updated without base year, 1991
yearL
```

```python
print(yearL[28])
print(len(yearL))
```

```python
for year in range(0,29): #for loop using index number, thus need to do last index of the list+1=length of the list
    #print(yearL[year])
    l=all_race.loc[all_race['DATA_YEAR']==yearL[year], "INCIDENT_MONTH"]
    for month in range(1,13):
        df=pd.DataFrame(columns=['year','month','hate_crime'])
        df.loc[month-1] = [yearL[year], month, sum(l==month)]
        all_race_ts = all_race_ts.append(df, ignore_index=True)
```

```python
all_race_ts #30 years x 12 months =360 rows
```

```python
outputfile1=os.path.join('/Users/mandyhong/Desktop', 'all_race_hc_ts.csv')
ts_df1=all_race_ts.to_csv(outputfile1, index=False)
```
