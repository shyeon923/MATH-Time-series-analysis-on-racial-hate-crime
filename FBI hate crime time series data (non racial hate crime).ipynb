### 4. for non-racial hate crime

```python
#make data that only has state, data_year, month left
non_race = non_raceHC[['DATA_YEAR', 'INCIDENT_MONTH']].copy()
non_race.index = range(len(non_race))
non_race
```

```python
#for each year, aggregate by month
#first, create 1991 data as base + test run for loop for month
yearL=pd.unique(non_race["DATA_YEAR"])
i=0
non_race_ts = pd.DataFrame(columns=['year','month','hate_crime']) #create data frame for each year with 12 months
l=non_race.loc[non_race['DATA_YEAR']==yearL[i], "INCIDENT_MONTH"]
#month=1
for month in range(1,13):
    non_race_ts.loc[month-1] = [yearL[i], month, sum(l==month)]
non_race_ts
```

```python
yearL=yearL[1:30] #list updated without base year, 1991
yearL
```

```python
print(len(yearL))
```

```python
for year in range(0,29): #for loop using index number, thus need to do last index of the list+1=length of the list
    #print(yearL[year])
    l=non_race.loc[non_race['DATA_YEAR']==yearL[year], "INCIDENT_MONTH"]
    for month in range(1,13):
        df=pd.DataFrame(columns=['year','month','hate_crime'])
        df.loc[month-1] = [yearL[year], month, sum(l==month)]
        non_race_ts = non_race_ts.append(df, ignore_index=True)
        
```
